```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_json('MyActivity.json',convert_dates=True)
```

```python
df = pd.DataFrame(df)
```

```python
df['time'] = df['time'].apply(pd.to_datetime)
```

```python
df = df[df.subtitles.notna()]
```

```python
def extract(x):
    return x[0]['name']
```

```python
names = map(extract,df.subtitles)
```

```python
df['names']=list(names)
```

```python
df['month'] = df.time.dt.strftime('%B %Y')
```

```python
df.names
```

```python
all_names=set(df.names)
```

```python
len(all_names)
```

```python
df=df.set_index('time')
```

```python
df11 = df.groupby(pd.Grouper(freq='M'))['names'].value_counts().groupby('names').cumsum()
```

```python
df11.max()
```

```python
df11_unstacked = df11.unstack(level=-1)
```

```python
df11_unstacked
```

```python
df11_unstacked = df11_unstacked.ffill()
```

```python
df11_unstacked.index.name = 'date'
```

```python
df12_unstacked = df11_unstacked.rank(axis=1,method='first',ascending=False)
```

```python
df11_unstacked.reset_index(inplace=True)
```

```python
df11_long = pd.melt(df11_unstacked,id_vars=['date'])
```

```python
df11_long[df11_long.value==225]
```

```python
df12_unstacked.reset_index(inplace=True)
```

```python
df12_long = pd.melt(df12_unstacked,id_vars=['date'])
```

```python
df11_long
```

```python
df12_long
```

```python
df11_long['rank']=df12_long['value']
```

```python
df11_long = df11_long.sort_values(by=['names','date'],ascending=False)
```

```python
df11_long
```

```python
df11_long['previous'] = df11_long.value.shift(periods=-1)
```

```python
df11_long['year'] = df11_long.date.dt.strftime('%Y')
```

```python
df11_long['month'] = df11_long.date.dt.strftime('%m')
```

```python
df11_long['year_month'] = (df11_long['year']) +'.'+ (df11_long['month'])
```
